```python
#maze.txt dosyamdan başlangıç ve bitiş pozisyonlarını oku, maze_size oluştur ve döndür
def load_maze(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    maze_size, start_position, end_position = map(tuple, (map(int, line.split()) for line in lines[:3]))
    maze = [list(line.strip()) for line in lines[3:]]
    return maze_size, start_position, end_position, maze

def print_maze(maze):
    for row in maze:
        print(''.join(row))
    print()

#başlangıç ve bitiş noktasına giden yolu hesapla .pop kullan
def solve_maze(maze_size, start_position, end_position, maze):
    rows, cols = maze_size
    stack = [(start_position, [])]
#.pop kullanma nedenim en son eklenen konum üzerinden ilerleyerek bir önceki konumdan devam etmek
    while stack:
        (current, path) = stack.pop()
        i, j = current

        if current == end_position:
            return path
#github'dan aldım algoyu
        if 0 <= i < rows and 0 <= j < cols and maze[i][j] == '.':
            #mevcut konumun boş bir yolda olduğunu kontrol eder
            maze[i][j] = '1'
            #aşağıda bulunan -1 +1 parametreleri  1 birim yukarı, aşağı, sola ve sağa gidilebilen konumlardır ve stacke eklenir
            stack.extend([
                ((i - 1, j), path + [(i - 1, j)]),
                ((i + 1, j), path + [(i + 1, j)]),
                ((i, j - 1), path + [(i, j - 1)]),
                ((i, j + 1), path + [(i, j + 1)])
            ])

#dosya konumunu oku
if __name__ == "__main__":
    maze_size, start_position, end_position, maze = load_maze(r'C:\Users\ASUS\Desktop\Maze Project Data\maze.txt')
    visualize_solution(maze_size, start_position, end_position, maze)
```
